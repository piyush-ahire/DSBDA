```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Load the Boston housing dataset
boston = fetch_openml(name="boston", as_frame=True)

# Convert the target variable to binary based on a threshold
threshold = boston.target.mean()  # You can choose any threshold based on your problem
boston.target = (boston.target > threshold).astype(int)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train the logistic regression model
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)

# Print predictions on the test data
y_pred = log_reg.predict(X_test_scaled)

# Convert the predictions to the format similar to linear regression output
print(y_pred.round(1))
```

```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import numpy as np

# Load the Boston housing dataset
boston = fetch_openml(name="boston", as_frame=True)

# Split the data into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(boston.data, boston.target, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train the linear regression model
reg = LinearRegression()
reg.fit(X_train_scaled, Y_train)

# Predict on the test set
y = reg.predict(X_test_scaled)

# Calculate Root Mean Squared Error
rmse = np.sqrt(mean_squared_error(Y_test, y))

# Calculate R^2
r2 = reg.score(X_test_scaled, Y_test)

print("The model performance for the test set")
print("--------------------------------------")
print("Root Mean Squared Error: {:.2f}".format(rmse))
print("R^2: {:.2f}".format(r2))
```
